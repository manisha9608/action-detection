### Create Train and Validation Data Generator objects

### Define and Compile Model

```python
model = model_gru_resnet.GRU_Resnet_Model()
model.compile(loss='sparse_categorical_crossentropy', run_eagerly = True, optimizer=tf.keras.optimizers.Adam(), metrics=['accuracy'])
```

### Train and Evaluate Model

```python
epochs = 1
hist_list = []
# model.save('saved_model')
for i in range(epochs):
    c=0
    for j in range(len(train_data)):
      if c<2:
        if c==0:
          c+=1
          model.fit_generator(train_data[j], workers=0)
          # hist_list.append(hist)
          model.save(model_path, save_format='tf')
          tf.keras.backend.clear_session()
          gc.collect()
          del model
        else:
          c+=1
          model = tf.keras.models.load_model(model_path)
          model.fit_generator(train_data[j], workers=0)
          # hist_list.append(hist)
          model.save(model_path, save_format='tf')
          tf.keras.backend.clear_session()
          gc.collect()
          del model
    val_loss = 0.0
    val_acc = 0.0
    # for k in range(len(val_data)):
    #     l, a = model.evaluate_generator(generator = val_data[k])
    #     val_loss += l
    #     val_acc += a
    # val_loss /= k
    # val_acc /=k
    
    # model.save('saved_model')
    tf.keras.backend.clear_session()
    # print("Epoch: ", i, ", Validation Loss: ",  val_loss, ", Validation Per-Frame Accuracy: ", val_acc)
```

### Test the model

```python

test_acc = 0.0
t=0
for i in range(len(test_data)):
  if t<1:
    t+=1
    model = tf.keras.models.load_model(model_path)
    _, a = model.evaluate_generator(test_data[i])
    test_acc += a
    model.save(model_path, save_format='tf')
    tf.keras.backend.clear_session()
    gc.collect()
    del model
test_acc/=i
print("Per Frame Accuracy for Test Data = ", test_ac)
```
